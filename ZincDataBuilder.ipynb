```python
def dataFrame_to_excel(dataFrame, pathIn, pathOut):
  df = pd.read_excel(pathIn)
  df = df.drop(0)
  df = df.rename(columns = {"Unnamed: 0" : "DateTime"})
  df = df[["DateTime", "Cu - AT502", "Cd - AT502"]]
  df = df.astype({col: float for col in df.columns[1:]})
  df['DateTime'] =  pd.to_datetime(df['DateTime'], format='%d.%m.%Y %H:%M:%S')
  df = df.set_index('DateTime')
  for i in df.index:
    if i in dataFrame.index: 
      df.at[i,"Cu - AT502"] = dataFrame.at[i, "CuOut"]
      df.at[i,"Cd - AT502"] = dataFrame.at[i, "CdOut"]
  
  df.to_excel(pathOut)
```
